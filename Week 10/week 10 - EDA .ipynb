# Data Cleaning

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:\\Users\\Shouq\\Downloads\\bank\\bank-full.csv')
df
```

```python
df.shape
```

```python
df.rename({'y':'deposit'},axis=1,inplace=True)
df
```

```python
df.info()
```

```python
df.describe()
```

## Checking the unique values 

```python
for col in df.columns:
    print(col)
    print(df[col].unique())
```

For job and martial columns we got some 'Unkown' values
to remove them first, we consider them as NA values and then use mode to replace the values 
since the proportion of unknown is considerably small

```python
df['job'].replace('unknown',np.nan,inplace = True)
df['marital'].replace('unknown',np.nan,inplace = True)
df
```

```python
df.isnull().sum()
```

Now job has 288 null values.

Next, we will be replacing NA values with the Mode.

```python
df['job'].replace(np.NAN , df['job'].mode()[0] , inplace=True)
df
```

```python
df.isnull().sum()
```

# Checking the outlier and skewness
